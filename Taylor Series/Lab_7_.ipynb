```python
import math

def func_e_to_3(n):
    x = 3
    e_to_3 = 0
    for i in range(n):
        e_to_3 += x**i/math.factorial(i)
    
    return e_to_3
```

```python
import math

def func_e(x, n):
    e_approx = 0
    for i in range(n):
        e_approx += x**i/math.factorial(i)
    
    return e_approx
```

```python
import math
def func_sin(x, n):
    sen_approx = 0
    for i in range(n):
        coef = (-1)**i
        num = x**(1+2*i)
        denom = math.factorial(1+2*i)
        sen_approx += ( coef ) * ( (num)/(denom) )
    
    return sen_approx
```

```python
angle_rad = (math.radians(60))
out = func_sin(angle_rad,7)
print(out)
```

```python
out = math.sin(angle_rad)
print(out)
```

- Reproduce TaylorSymPy (based on Taylor Series Expansion with Python from Data Science Fabric) for the functions in the Taylor Series figure above.

```python
from sympy import series, Symbol
from sympy.functions import sin, cos, exp,log
from sympy.plotting import plot
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = 10,10
plt.rcParams['lines.linewidth'] = 2
```

```python
# Define symbol
x = Symbol('x')
```

```python
# Function for Taylor Series Expansion

def taylor(function, x0, n):
    """
    Parameter "function" is our function which we want to approximate
    "x0" is the point where to approximate
    "n" is the order of approximation
    """
    return function.series(x,x0,n).removeO()
```

```python
print('sin(x) =', taylor(sin(x), 0, 4))

print('cos(x) =', taylor(cos(x), 0, 4))

print('e(x) =', taylor(exp(x), 0, 4))
```

```python
import math

def func_ln1x(x, n):
    ln_approx = 0
    for i in range(1,n):
        coef = (-1)**(i-1)
        num = x**(i)
        denom = i
        ln_approx += ( coef ) * ( (num)/(denom) )
    
    return ln_approx
```

```python
import math

x = .2
epsilon = 10**(-10)

for k in range(1000):
  if (  ( (x**(k+1)) / math.factorial(k+1) ) < epsilon):
    break
print(k)

ln_x = 0
for i in range(1,k):
    print(i,' ',ln_x, ' delta ', abs((-1)**(i-1)*(x**(i)) / (i)))
    ln_x += (-1)**(i-1)*(x**(i)) / (i)
    
print('math.log(1.5) = ',math.log(1+x))
```

```python
x = .2
epsilon = 10**(-10)
delta = 1
ln_x = 0
i = 1

while  epsilon <= delta:
  delta = abs((-1)**(i-1)*(x**(i)) / (i))
  if i%2 == 0:
    ln_x -= delta
  else:
    ln_x += delta
  i = i+1
print(i-1) 
print('ln_x = ',ln_x)   
    
print('math.sin(1.2) = ',math.log(1+x))
```

```python
import math

x = (math.radians(60))
epsilon = 10**(-20)
delta = 1
sen_x = 0
i = 0

while  epsilon <= delta:
  delta = abs((-1)**i*(x**(1+2*i)) / math.factorial(1+2*i))
  if i%2 == 0:
    sen_x += delta
  else:
    sen_x -= delta
  i = i+1
print(i-1) 
print('sen_x = ',sen_x)   
    
print('math.sin(60) = ',math.sin(x))
```
